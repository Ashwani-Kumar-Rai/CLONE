```python
# DISPLAYING VIDEO CAPTURE DEVICE
import numpy as np 
import cv2 

# cv2.VideoCapture(0) : this will take hold of the camera 
cap = cv2.VideoCapture(0) # if you have multiple video cameras then 0,1,2,3 for all those camera
# if you want to load from a file then ('filename')

# this while loop will run till we press a key on our keyboard to quit , till key displaying the video device 
while True:
    # get a frame from our video capture device  , cap.read() will return 2 things 
    ret ,frame = cap.read()  # cap.read() will [frame]return the image itself[ i.e numpy array] 
                                #- ret will tell you if the capture works properly or not 

# display the frame
    cv2.imshow('frame',frame)

    if cv2.waitKey(1) == ord('q'): # if we press key within 1 milli second ,cv2.waitKey(1) will return the ordinal 
                                        #- value or the ascii value associated with the key that we pressed 
        break 

cap.release() # this will release the camera resource , so another program like obs may use the camera    

# destroyAllWindows() function allows users to destroy or close all windows at any time after exiting the script.
cv2.destroyAllWindows()


# WORKING 
# the way this works , we are going through all of the code here , and then every single frame ,
    #- so every single millisecond ,it is grabbing/reading information from the camera and then displaying
        #- then on the screan 
```

```python
# MIRROR VIDEOS MULTIPLE TIMES 

# How can we maipulate that so that there is four on the screen , then rotate all of them in different directions 
# i want to take this frame here , rather then showing this image , i want to turn this image into 4 seperate imagecap = cv2.VideoCapture(0) # if you have multiple video cameras then 0,1,2,3 for all those camera
import numpy as np 
import cv2 

cap = cv2.VideoCapture(0)

while True:
    ret ,frame = cap.read()  

    # we need to find the width and height of the frame , and at the quadrant we will place our images 
    width = int(cap.get(3)) # cap.get(3) : this will give the width property of your frame 
                            #- 3 is the identifier for a property , in this case that property is width 

    height = int(cap.get(4)) # 4 is height

    # create a blank canvas so that we can put our face into it 
    image = np.zeros(frame.shape,np.uint8) # pass the shape of the image
    smaller_frame= cv2.resize(frame,(0,0),fx=0.5,fy=0.5)

    # we are taking the smaller frame and we are pasting it into the top ledt hand corner of the new image. that we just created 
    image[:height//2 , :width//2] = cv2.rotate(smaller_frame,cv2.ROTATE_180)# top left 
    image[height//2: , :width//2] = smaller_frame # bottom left 
    image[:height//2 , width//2:] = cv2.rotate(smaller_frame,cv2.ROTATE_180)# # top right
    image[height//2: , width//2:] = smaller_frame # bottom right
    # to rotate this image , i just have to rotate this smaller frame 



    cv2.imshow('frame',image) 
    if cv2.waitKey(1) == ord('q'):                       
        break 

cap.release() 
cv2.destroyAllWindows()


```
